```python
import glycosylator as gl
from glob import glob
```

```python
rev_names = {}
for image in glob("*.png"):
    if image == "unknown.png":
        continue
    name = image.split(".")[0]
    if not gl.has_compound(name):
        continue
    _dict = gl.get_compound(name, return_type="dict")
    if isinstance(_dict, list):
        for i in _dict:
            i = i[0]["id"]
            rev_names[i] = name
    else:
        rev_names[_dict[0]["id"]] = name

    name = "b-" + name
    if not gl.has_compound(name):
        continue
    _dict = gl.get_compound(name, return_type="dict")
    if isinstance(_dict, list):
        for i in _dict:
            i = i[0]["id"]
            rev_names[i] = name
    else:
        rev_names[_dict[0]["id"]] = name

rev_names
```
